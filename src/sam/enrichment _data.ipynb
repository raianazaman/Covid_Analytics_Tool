<font size = 5>Merge Enrichment Data</font>

```python
import pandas as pd
import numpy as np
```

<font size="5">Load Data</font>

```python
#read the super data
superdata = pd.read_csv("..\..\data\superCOVID-19datafame.csv")
#fill zeros to make countyFIPS five digits
superdata['countyFIPS'] = superdata.countyFIPS.astype('str')
superdata['countyFIPS'] = superdata['countyFIPS'].apply(lambda x: x.zfill(5))
#display dataframes
superdata.head(10)
```

```python
#read the Census Demographic ACS data
enrichdata = pd.read_csv("..\..\data\modified_DP.csv")
#display dataframes
enrichdata.head(10)
```

<font size="5">Merge</font>

```python
#take last five digits of column id as countyFIPS
enrichdata['countyFIPS'] = enrichdata['id'].str[-5:]
#rearrange columns
cols = enrichdata.columns.tolist()
cols = cols[-1:] + cols[:-1]
enrichdata = enrichdata[cols]
enrichdata.head()
```

```python
#since the enrichdata doesn't have all the counties, select the county it has from superdata
smalldata = superdata[superdata.set_index(['countyFIPS']).index.isin(enrichdata.set_index(['countyFIPS']).index)]
smalldata.head()
```

```python
#set index for merging
enrichdata.set_index(['countyFIPS'])
#merge dataframes
merged = pd.merge(smalldata, enrichdata, left_index=True, right_index=True)
merged.head(10)
```

```python
#Save to data folder 
merged.to_csv("enrichsuperdata.csv")
```
