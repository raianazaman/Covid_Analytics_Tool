```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

<font size="5">Load Data</font>

```python
#read the super data
superdata = pd.read_csv("..\..\data\superCOVID-19datafame.csv")
#display dataframes
superdata.head(10)
```

<font size="5">Modify and Analyze data</font>

```python
#select data where state is IN
indiana = superdata.loc[superdata['State'].isin(['IN'])]
indiana.to_csv("indiana_data.csv")
indiana.head()
```

```python
#select data from the last week
weekdata = indiana[["countyFIPS", "County Name", "State", "confirmed_9/13/20", "confirmed_9/14/20", "confirmed_9/15/20", "confirmed_9/16/20", "confirmed_9/17/20","confirmed_9/18/20","confirmed_9/19/20"]]
weekdata.head(10)
```

```python
#show the total sum of confirmed cases in each day
dailycase = weekdata[["confirmed_9/13/20","confirmed_9/14/20","confirmed_9/15/20","confirmed_9/16/20","confirmed_9/17/20"
,"confirmed_9/18/20","confirmed_9/19/20"]].sum()
dailycase.head(7)
```

As we can see, the confirmed case in Indiana is increasing during the last week in report.

```python
#plot the confirmed cases
dailycase.plot()
```
