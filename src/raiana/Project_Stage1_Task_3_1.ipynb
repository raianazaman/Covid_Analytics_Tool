#Reading Super Dataset

```
import pandas as pd
import numpy as np

#READING THE superCOVID_dataset FILE
superCOVID_dataset = pd.read_csv("/content/drive/My Drive/Datascience/Project/COVID-Team-2/data/superCOVID-19datafame.csv")
superCOVID_dataset.head(10)


```

#Showing Data for Newyork only

```

#remove county fips with zeros
team_super_NYC = superCOVID_dataset[superCOVID_dataset.apply(lambda x: x["countyFIPS"] != 0, axis=1)]

# selecting rows based on condition 
options = ['NY']  
#filter data download Newyork only
team_super_NYC = superCOVID_dataset[superCOVID_dataset['State'].isin(options)] 

team_super_NYC.to_csv("raiana.nyc.csv")
team_super_NYC.head()
```

#Reading Data for a week

```

#Reading Data for a week
nycWeeklyReport = team_super_NYC[["countyFIPS", "County Name", "State"
,"confirmed_9/13/20","confirmed_9/14/20","confirmed_9/15/20","confirmed_9/16/20","confirmed_9/17/20","confirmed_9/18/20","confirmed_9/19/20"
,"deaths_9/13/20","deaths_9/14/20","deaths_9/15/20","deaths_9/16/20","deaths_9/17/20","deaths_9/18/20","deaths_9/19/20"]]

nycWeeklyReport.to_csv("raiana_nycWeeklyReport.csv")
nycWeeklyReport.head(10) 
```

```
#Sum of the Confirmd case in Nyc in last week
sum_row=nycWeeklyReport[["confirmed_9/13/20","confirmed_9/14/20","confirmed_9/15/20","confirmed_9/16/20","confirmed_9/17/20"
,"confirmed_9/18/20","confirmed_9/19/20"]].sum()

sum_row
```

```
#Sum of the Death case in Nyc in last week
sum_Death=nycWeeklyReport[["deaths_9/13/20","deaths_9/14/20","deaths_9/15/20","deaths_9/16/20",
                           "deaths_9/17/20","deaths_9/18/20","deaths_9/19/20"]].sum()

sum_Death
```
