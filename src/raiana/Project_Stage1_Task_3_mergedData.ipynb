#Reading Super Dataset

```
import pandas as pd
import numpy as np

#READING THE superCOVID_dataset FILE
super_COVID_dataset = pd.read_csv("/content/drive/My Drive/Datascience/Project/COVID-Team-2/data/superCOVID-19datafame.csv")
print(super_COVID_dataset)

HospitalBed_dataset = pd.read_csv("/content/drive/My Drive/Datascience/Project/COVID-Team-2/data/Definitive_Health_USA_Hospital_Beds.csv")
print(HospitalBed_dataset)

```

#Cleaninging Data 

```

#remove county fips with zeros
super_COVID_dataset = super_COVID_dataset[super_COVID_dataset.apply(lambda x: x["countyFIPS"] != 0, axis=1)]

#Make countyFIPS column a number
super_COVID_dataset["countyFIPS"] = super_COVID_dataset["countyFIPS"].apply(pd.to_numeric)


#Reindex super file now that we have removed the unneed counties
super_COVID_dataset.set_index("countyFIPS", inplace=True)
```

```
#remove county fips with zeros
HospitalBed_dataset = HospitalBed_dataset[HospitalBed_dataset.apply(lambda x: x["FIPS"] != 0, axis=1)]

#Make countyFIPS column a number
HospitalBed_dataset["FIPS"] = HospitalBed_dataset["FIPS"].apply(pd.to_numeric)

#Reindex super file now that we have removed the unneed counties
HospitalBed_dataset.set_index("FIPS", inplace=True)
```

```

HospitalBed_dataset_clean = HospitalBed_dataset[["FIPS", "HQ_STATE","NUM_STAFFED_BEDS","NUM_LICENSED_BEDS","NUM_ICU_BEDS"]]


HospitalBed_dataset_clean.head(10) 
```

```
#converting FlIP from Float to int 
HospitalBed_dataset_clean['FIPS'] = HospitalBed_dataset_clean['FIPS'].fillna(0).astype(np.int64)
HospitalBed_dataset_clean['FIPS'] = HospitalBed_dataset_clean['FIPS'].astype(np.int64)
HospitalBed_dataset_clean.set_index("FIPS",inplace=True)
HospitalBed_dataset_clean.head()
```

#Merge super file with HospitalBed dataset

```

merged=pd.merge(left=super_COVID_dataset, right=HospitalBed_dataset_clean, left_on='countyFIPS', right_on='FIPS', how='inner')
merged.head(20)
```

```
merged.to_csv("raiana.mergedData.csv")
```
