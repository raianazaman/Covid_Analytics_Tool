```python
import pandas as pd
import numpy as np
import statsmodels.formula.api as smf
from sklearn.model_selection import train_test_split
from statsmodels.tools.eval_measures import rmse
from util import *
import matplotlib.pyplot as plt
```

```python
data_path = "../../data/"
overall = pd.read_csv(data_path + "superCOVID-19dataframe.csv", index_col = "countyFIPS").drop(0)
county_names = overall['County Name']
overall.drop(columns = ['County Name'], inplace=True)
overall = per_N(overall, 1000000)
```

```python
def get_known_dfs(frame):
    cases = get_cases(frame, 0, True).sum().to_frame()
    deaths = get_deaths(frame, 0, True).sum().to_frame()
    cases.index = range(len(cases))
    deaths.index = range(len(deaths))
    cases.reset_index(inplace=True)
    deaths.reset_index(inplace=True)
    cases.rename(columns = {'index': 'day', 0: 'total'}, inplace=True)
    deaths.rename(columns = {'index': 'day', 0: 'total'}, inplace=True)
    return cases, deaths    
```

```python
cases, deaths = get_known_dfs(overall)
cases.head()
```

```python
poly_1 = smf.ols(formula=" total ~ day", data=cases).fit()
```

```python
days = cases['day']
pred = poly_1.predict(days)
pred.head()
```

```python
tmp = pd.DataFrame({'day': [days.min(), days.max()]})
line =  poly_1.predict(tmp) 
```

```python
cases.plot(kind="scatter", x='day', y='total')
plt.plot(tmp, line, c="red", linewidth = 2)
```

```python
plt.figure(figsize=(8 * 1.618, 8))

plt.scatter(cases.day, cases.total)

plt.plot(cases.day, poly_1.predict(days), 'b-', label='Poly n=1 $R^2$=%.2f' % poly_1.rsquared, alpha=0.9)

# 2-nd order polynomial
poly_2 = smf.ols(formula='total ~ 1 + day + I(day ** 2.0)', data=cases).fit()
plt.plot(cases.day, poly_2.predict(days), 'g-', label='Poly n=2 $R^2$=%.2f' % poly_2.rsquared, alpha=0.9)

# 3-rd order polynomial
poly_3 = smf.ols(formula='total ~ 1 + day + I(day ** 2.0) + I(day ** 3.0)', data=cases).fit()
plt.plot(cases.day, poly_3.predict(days), 'r-', alpha=0.9, label='Poly n=3 $R^2$=%.2f' % poly_3.rsquared)

legend = plt.legend()
```

```python
train, test = train_test_split(cases)
```

```python
poly_3 = smf.ols(formula='total ~ 1 + day + I(day ** 2.0) + I(day ** 3.0)', data=train).fit()
```

```python
test_preds1 = poly_1.predict(test['day'])
test_preds2 = poly_2.predict(test['day'])
test_preds3 = poly_3.predict(test['day'])
```

```python
rmse1 = rmse(test_preds1, test['total'])
rmse2 = rmse(test_preds2, test['total'])
rmse3 = rmse(test_preds3, test['total'])
print(rmse1, rmse2, rmse3)
```

Identify which counties are most at risk

```python
top5 = [13121, 13135, 13067, 13089, 13139]
counties = overall.loc[top5]
counties.head()
```

```python
for county in counties.index:
    c,d = get_known_dfs(overall.loc[[county]])
    errs = d.loc[d['total'] < 0].index
#     print(errs)
    d.drop(errs, inplace=True)    
    poly_3c = smf.ols(formula='total ~ 1 + day + I(day ** 2.0) + I(day ** 3.0)', data=c).fit()
    poly_3d = smf.ols(formula='total ~ 1 + day + I(day ** 2.0) + I(day ** 3.0)', data=d).fit()

    fig, axs = plt.subplots(1,2, figsize=(12,6))
    axs[0].scatter(c.day, c.total)
    axs[0].plot(c.day, poly_3c.predict(c.day), 'r-', alpha=0.9, label='Poly n=3 $R^2$=%.2f' % poly_3c.rsquared)
    axs[1].scatter(d.day, d.total, c='r')
    axs[1].plot(d.day, poly_3d.predict(d.day), 'g-', alpha=0.9, label='Poly n=3 $R^2$=%.2f' % poly_3d.rsquared)    
```
